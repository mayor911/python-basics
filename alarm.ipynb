```python
from os import system as sys
import pygame
from time import sleep
```

```python
def get_time():
    sys('date > date')
    
    date_file = open('date', 'r')
    date = date_file.readline()
    parts = date.split()
    date = parts[3]
    hour = date.split(':')[0]
    minute = date.split(':')[1]
    second = date.split(':')[2]
    return {'Hour': hour, 'Minute': minute}

def set_time():
    time = input('Set time format (10:10) ')
    parts = time.split(':')
    return {'Hour': parts[0], 'Minute': parts[1]}

def play_audio(filename):
    pygame.mixer.init()
    sound = pygame.mixer.Sound(filename)
    sound.play()
    sleep(20)
    pygame.mixer.music.stop()
    del sound

print(get_time())
time = set_time()
while time != get_time():
    print('.', end='')
    sleep(1)


play_audio('Tangled_up_lutanfyah.mp3')
```
